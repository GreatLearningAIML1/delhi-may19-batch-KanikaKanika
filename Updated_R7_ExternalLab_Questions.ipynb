### Build a DNN using Keras with `RELU` and `ADAM`

#### Load tensorflow

```python
import tensorflow  as tf
from keras.utils import np_utils
```

#### Collect Fashion mnist data from tf.keras.datasets 

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

#### Change train and test labels into one-hot vectors

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

#### Initialize model, reshape & normalize data

```python
TRAIN = False
BATCH_SIZE = 32
EPOCHS = 100
```

#### Add two fully connected layers with 200 and 100 neurons respectively with `relu` activations. Add a dropout layer with `p=0.25`

```python
model1 = tf.keras.models.Sequential()
```

```python
model1.add(tf.keras.layers.Reshape((784,),input_shape=(28,28,)))
```

```python
#First Fully Connected Layer
model1.add(tf.keras.layers.Dense(200,activation="relu"))

#Second Fully Connected Layer
model1.add(tf.keras.layers.Dense(100,activation="relu"))

#Drop out Layer
model1.add(tf.keras.layers.Dropout(0.25))
```

### Add the output layer with a fully connected layer with 10 neurons with `softmax` activation. Use `categorical_crossentropy` loss and `adam` optimizer and train the network. And, report the final validation.

```python
#Output Layer
model1.add(tf.keras.layers.Dense(10,activation="softmax"))
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model1.fit(x_train, y_train, 
          validation_data=(x_test, y_test), 
          epochs=EPOCHS,
          batch_size=BATCH_SIZE)
```

# Creating a new model by adding the Convolutional Layer and forming fully connected network and then checking the accuracy

```python
import keras
from keras.layers import Dense, Activation, Dropout, Flatten, Reshape
from keras.layers import Convolution2D, MaxPooling2D
```

```python
#Reshaping the X_train and X_test
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

# Define model
model2 = Sequential()

# 1st Conv Layer
model2.add(Convolution2D(32, 3, 3, input_shape=(28, 28, 1)))
model2.add(Activation('relu'))

# 2nd Conv Layer
model2.add(Convolution2D(32, 3, 3))
model2.add(Activation('relu'))

```

```python
#Fully Connected Layer 1
model2.add(Flatten())
model2.add(Dense(200))
model2.add(Activation('relu'))

#Fully Connected Layer 2
model2.add(Dense(100))
model2.add(Activation('relu'))
    
```

```python
#Add Dropout
model2.add(Dropout(0.25))

#Prediction Layer
model2.add(Dense(10))
model2.add(Activation('softmax'))

```

```python
# Loss and Optimizer
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
# Store Training Results
early_stopping = keras.callbacks.EarlyStopping(monitor='val_acc', patience=7, verbose=1, mode='auto')
callback_list = [early_stopping]

# Train the model2
model2.fit(x_train, y_train, batch_size=BATCH_SIZE, nb_epoch=10, 
              validation_data=(x_test, y_test),callbacks=callback_list)
```
